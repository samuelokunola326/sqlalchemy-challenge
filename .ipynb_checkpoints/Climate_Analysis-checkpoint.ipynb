# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect, func
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Integer, String, Float

Base = declarative_base()
```

```python
#Connecting to the database 
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
#reflect database into new model
Base = automap_base()

#reflecting tables
Base.prepare(engine,reflect=True)

# Print all of the classes mapped to the Base
Base.classes.keys()
```

```python
#using inspector to get column names and info
inspector = inspect(engine)
m_columns = inspector.get_columns('measurement')
m_columns
```

```python
#using inspector to get column names and info
s_columns = inspector.get_columns('station')
s_columns
```

```python
#reflect the tables
measurements = Base.classes.measurement
station = Base.classes.station
```

```python
#creating session
session = Session(engine)
```

# Exploratory Climate Analysis

```python
#Getting max date for later query
engine.execute('SELECT MAX(date) FROM measurement').fetchall()
```

```python
#Creating a query that filters the last 12 months using above max date
prcp_data = session.query(measurements.date,
                          measurements.prcp ).filter(measurements.date <= '2017-08-23',measurements.date >= '2016-08-23').all()

prcp_data
```

```python
#Reading data into Pandas
hw_df = pd.DataFrame(prcp_data)
#showing data frame
hw_df
```

```python
#sorting data frame
hw_df = hw_df.sort_values(by='date')
#resetting index
hw_df.reset_index(drop=True)
```

```python

#Plotting prcp data
hw_df_plot = hw_df.plot(x='date',y='prcp', rot=90, figsize=(20,10))

#Adding title for chart
plt.title("Precipitation from Aug 2016 to Aug 2017")

#labeling y label
plt.ylabel('Inches')

#showing chart
plt.show()
```

```python
#showing summary stats
hw_df.describe()
```

```python
#query to get unique station names and how many of them 
station_data = session.query(station.name).distinct()

# empty list to hold unique station names
station_list = []
#looping through the results of query
for i in station_data:
    station_list.append(i)

#Printing the number of stations and names
print(len(station_list))
print(station_list)
```

```python
#query to show most active stations in desc order
active_stations = session.query(measurements.station,
                          func.count(measurements.station)).group_by(measurements.station).order_by(func.count(measurements.station).desc()).all()
active_stations
```

```python
#Query to get name of most active station name
sel = [measurements.station, station.name]
most_active_station = session.query(*sel).filter(measurements.station == station.station, 
                                                 measurements.station =='USC00519281' ).limit(1).all()
#showing results
most_active_station
```

```python
# fetching data from data base
tobs_data = session.query(measurements.tobs).filter(measurements.date <= '2017-08-23',
                                               measurements.date >= '2016-08-23', measurements.station =='USC00519281').all()
```

```python
# creating data frame to plot data
tobs_df = pd.DataFrame(tobs_data)
tobs_df
```

```python
#Plotting prcp data
tobs_df_plot = tobs_df['tobs'].plot.hist(bins=12, rot=360, figsize=(20,10))

#showing legends
plt.legend()

#Adding title for chart
plt.title("Temp Frequency from Aug 2016 to Aug 2017")

#labeling y label
plt.ylabel('Frequency')
plt.xlabel('Temperature')

#showing chart
plt.show()
```

```python
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()

# function usage example
print(calc_temps('2012-02-28', '2012-03-05'))
```

```python
# Create a query that will calculate the daily normals 
# (i.e. the averages for tmin, tmax, and tavg for all historic data matching a specific month and day)

def daily_normals(date):
    """Daily Normals.
    
    Args:
        date (str): A date string in the format '%m-%d'
        
    Returns:
        A list of tuples containing the daily normals, tmin, tavg, and tmax
    
    """
    
    sel = [func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)]
    return session.query(*sel).filter(func.strftime("%m-%d", Measurement.date) == date).all()
    
daily_normals("01-01")
```
